```python
Architecture_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Architecture,Building and Planning"')
Arts_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Arts"')
Business_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Business and Economics"')
Engineering_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Engineering"')
Fine_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Fine Arts and Music"')
Law_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Law"')
Medicine_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Medicine, Dentistry and Health Sciences"')
BusinessSchool_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Melbourne Business School"')
GraduateSchoolEducation_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Melbourne Graduate School of Education"')
ProfessioinalContiuingEducation_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Melbourbe School of Professioinal and Contiuing Education"')
Science_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Science"')
StudentScholarlyServices_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Student and Scholarly Services"')
Commercial_df = PG_df.query('AVAIL_ORG_UNIT_NM == "UoM Commercial Ltd"')
VCA_MCM_df = PG_df.query('AVAIL_ORG_UNIT_NM == "VCA and MCM"')
Veterinary_Agricultural_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Veterinary & Agricultural Sicences"')
VeterinaryAndAgricultural_df = PG_df.query('AVAIL_ORG_UNIT_NM == "Veterinary and Agricultural Sicences"')
```

```python
print("The dimension of Architecture,Building and Planning:",Architecture_df.shape)
print("The dimension of Arts",Arts_df.shape)
print("The dimension of Business and Economics:",Business_df.shape)
print("The dimension of Engineering",Engineering_df.shape)
print("The dimension of Fine Arts and Music:",Fine_df.shape)
print("The dimension of Law",Law_df.shape)
print("The dimension of Medicine, Dentistry and Health Sciences",Medicine_df.shape)
print("The dimension of Melbourne Business School:",BusinessSchool_df.shape)
print("The dimension of Melbourne Graduate School of Education:",GraduateSchoolEducation_df.shape)
print("The dimension of Melbourbe School of Professioinal and Contiuing Education:",ProfessioinalContiuingEducation_df.shape)
print("The dimension of Science:",Science_df.shape)
print("The dimension of Student and Scholarly Services:",StudentScholarlyServices_df.shape)
print("The dimension of UoM Commercial Ltd:",Commercial_df.shape)
print("The dimension of VCA and MCM:",VCA_MCM_df.shape)
print("The dimension of Veterinary & Agricultural Sicences:",Veterinary_Agricultural_df.shape)
print("The dimension of Veterinary and Agricultural Sicences:",VeterinaryAndAgricultural_df.shape)
```
