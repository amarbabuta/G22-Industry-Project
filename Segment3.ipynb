```python
def data_split_to_seconds(dat):
    d = str.split(dat,' ')
    if len(d) == 3:
        day_sec = int(d[0]) * 86400
        time = d[2].split(':')
        time_sec = int(time[0]) * 3600 + int(time[1]) * 60 + int(time[2])
        total_sec = day_sec + time_sec
        return total_sec
    else:
        return 0
```

```python
wrkflw33 = pd.read_csv('processtime3.csv')
```

```python
wrkflw33['processtime_sec'] = wrkflw33['processtime3'].apply(data_split_to_seconds)
```

```python
seconds3 = wrkflw33['processtime_sec'].mean()
seconds3
```

```python
med_seconds = wrkflw33['processtime_sec'].median()
med_seconds
```
