```python
def data_split_to_seconds(dat):
    d = str.split(dat,' ')
    if len(d) == 3:
        day_sec = int(d[0]) * 86400
        time = d[2].split(':')
        time_sec = int(time[0]) * 3600 + int(time[1]) * 60 + int(time[2])
        total_sec = day_sec + time_sec
        return total_sec
    else:
        return 0
```

```python
wrkflw22['processtime_sec'] = wrkflw22['processtime2'].apply(data_split_to_seconds)
```
