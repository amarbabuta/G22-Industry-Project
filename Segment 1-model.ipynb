```python
import csv
import pandas as pd
from sklearn.linear_model import LogisticRegression 
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.svm import LinearSVC
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
```

```python
segment1=pd.read_csv('segment1_clean.csv')
```

```python
del segment1['Unnamed: 0']
```

```python
segment1.dtypes
```

```python
segment1.head(30)
```

```python
seg1=segment1[['CASE_ID','STU_CITIZEN_NM','STU_ABOR_TSI_NM','SUBMISSION_METHOD','ENTRY_TYPE','LAST_COUNTRY_EDUCATION',
          'PREFERENCE_NO','AVAIL_ORG_UNIT_NM','STUDY_MODE_CD','ORIG_AVAIL_YR','ORIG_STUDY_PERIOD','SPK_CAT_TYPE_CD',
          'OFFER_RELEASED_FG']]
```

```python
seg1.dtypes
```

#### Labelling the categorical variables

```python
seg1.iloc[:,3] = LE.fit_transform(seg1.iloc[:,3].astype(str))
seg1.iloc[:,4] = LE.fit_transform(seg1.iloc[:,4].astype(str))
seg1.iloc[:,5] = LE.fit_transform(seg1.iloc[:,5].astype(str))
seg1.iloc[:,12] = LE.fit_transform(seg1.iloc[:,12].astype(str))
```

```python
seg1.iloc[:,1] = LE.fit_transform(seg1.iloc[:, 1].values)

seg1.iloc[:,2] = LE.fit_transform(seg1.iloc[:, 2].values)

seg1.iloc[:,7] = LE.fit_transform(seg1.iloc[:, 7].values)

seg1.iloc[:,8] = LE.fit_transform(seg1.iloc[:, 8].values)

seg1.iloc[:,10] = LE.fit_transform(seg1.iloc[:, 10].values)

seg1.iloc[:,11] = LE.fit_transform(seg1.iloc[:, 11].values)
```

```python
seg1.dtypes
```

#### For Response Variable 1 (OFFER_RELEASED_FG())

```python
X=seg1.iloc[:,1:12].values
Y=seg1.iloc[:,12].values 
```

```python
#spliting dataset
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size = 0.3,random_state=0)
```

```python
# Logistic Regression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_predict=lr.predict(x_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, lr_predict)*100,2)))
```

```python
# Confusion matrix, f1_sccore, Precision, Recall
score=accuracy_score(y_test,lr_predict)
print(f'Accuracy: {round(score*100,2)}%')

cm=confusion_matrix(y_test,lr_predict, labels=[1,0])
print ("\n Confusion Matrix : \n")
print (cm)

p, r, f, _ = precision_recall_fscore_support(y_test,lr_predict, pos_label=1, average="micro");

print("\nPerformance on offer released:")
print("Precision =", p)
print("Recall    =", r)
print("F1        =", f)
```

```python
# Linear SVM
svm = LinearSVC()
svm.fit(x_train, y_train)
```

```python
svm_predict = svm.predict(x_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, svm_predict)*100,2)))
```

```python
# K Nearest Neighbors
Knn = KNeighborsClassifier()
Knn.fit(x_train,y_train)
```

```python
knn_predict=Knn.predict(x_test)
print("accuracy: {}%".format(round(accuracy_score(y_test, knn_predict)*100,2)))
```
