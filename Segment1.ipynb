```python
def data_split_to_seconds(dat):
    d = str.split(dat,' ')
    if len(d) == 3:
        day_sec = int(d[0]) * 86400
        time = d[2].split(':')
        time_sec = int(time[0]) * 3600 + int(time[1]) * 60 + int(time[2])
        total_sec = day_sec + time_sec
        return total_sec
    else:
        return 0
```

```python
wrkflw = pd.read_csv('processtime1.csv')
```

```python
wrkflw['processtime_sec'] = wrkflw['processtime1'].apply(data_split_to_seconds)
```

```python
seconds = wrkflw['processtime_sec'].mean()
seconds
```

```python
med_seconds = wrkflw['processtime_sec'].median()
med_seconds
```

```python
days = seconds // 86400
hours = (seconds - (days * 86400)) // 3600
minutes = (seconds - (days * 86400) - (hours * 3600)) // 60
print(days, hours, minutes)
```

```python
med_days = med_seconds // 86400
med_hours = (med_seconds - (med_days * 86400)) // 3600
med_minutes = (med_seconds - (med_days * 86400) - (med_hours * 3600)) // 60
print(med_days, med_hours, med_minutes)
```
